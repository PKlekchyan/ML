#### Проведем кластеризацию, попробуем разбить на отдельные районы

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, AffinityPropagation
from sklearn.metrics import adjusted_rand_score, silhouette_score
from sklearn.preprocessing import StandardScaler 
```

```python
scaler = StandardScaler()
```

```python
x = scaler.fit_transform(df[['x', 'y']])
```

#### K-Means

```python
k_inertia = []
ks = range(1,15)

for k in ks:
    model_kmeans = KMeans(n_clusters=k)
    clusters_kmeans = model_kmeans.fit_predict(x)
    k_inertia.append(model_kmeans.inertia_)
```

```python
plt.plot(ks, k_inertia)
plt.xlabel('n_clusters')
plt.grid()
plt.show()
```

```python
diff = np.diff(k_inertia)
```

```python
plt.plot(ks[1:], diff)
plt.grid()
plt.show()
```

```python
diff_r = diff[1:] / diff[:-1]
```

```python
plt.plot(ks[1:-1], diff_r)
plt.grid()
plt.show()
```

```python
model_kmeans = KMeans(n_clusters=8)
clusters_kmeans = model_kmeans.fit_predict(x)
```

```python
# plt.figure(figsize=(10, 8))
plt.scatter(x[:, 0], x[:, 1], c=clusters_kmeans, s=60)
plt.show()
```

```python
model_dis_kmeans = KMeans(n_clusters = 8)
clusters_dis_kmeans = model_agr_kmeans.fit_predict(x_disagree)
```

### Основные тематики комментариев 

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
import pymorphy2
import re
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
data = df[['multiplier', 'comment']]
data.head()
```

```python
morph = pymorphy2.MorphAnalyzer()
stop_words = stopwords.words('russian')
```

#### Приводим к лемме

```python
clean_dict = {}
counter = 0

for num, comment in zip(range(len(data)), data['comment']):
    all_tokens = []
    sentences = nltk.sent_tokenize(comment, language='russian')
    for sent in sentences:
        sent_ = re.sub(r'[^А-Яа-яёЁ ]', '', sent)
        tokens = nltk.word_tokenize(sent_)
        tokens = [i for i in tokens if (i not in stop_words)]
        tokens = list(map(lambda x:morph.parse(x)[0].normal_form, tokens))
        all_tokens.append(tokens)
    clean_dict[num] = all_tokens
    
    counter += 1
    if counter % 500 == 0:
        print("{0} docs processed".format(counter))
#     if counter == 50000: # для простоты расчета возьмем выборку из 50 000
#         break
```

```python
def get_arr(dict_):
    arr = []
    for key, value in dict_.items():
        help_arr = []
        for i in value:
            help_arr +=i
        arr.append(help_arr)
    return arr
```

```python
arr = get_arr(clean_dict)
```

```python
# Gensim
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel

# # Plotting tools
# import pyLDAvis
# import pyLDAvis.gensim  # don't skip this

import warnings
warnings.filterwarnings("ignore",category=DeprecationWarning)
```

#### Создаем dict и corpus из gensim

```python
id2word = corpora.Dictionary(arr)
```

```python
corpus = [id2word.doc2bow(i) for i in arr]
```

#### Построение Тематической модели

```python
import pyLDAvis
# import pyLDAvis.gensim  # don't skip this
```

```python
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=5, 
                                           random_state=100,
                                           update_every=1,
                                           chunksize=5000,
                                           passes=10,
                                           alpha='auto',
                                           per_word_topics=True)
```
